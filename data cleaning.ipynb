```python
 import pandas as pd
```

```python
df= pd.read_excel(r'E:\datasets\Customer Call List.xlsx')
df
```

```python
df.drop_duplicates()
```

```python
df.drop(columns ='Not_Useful_Column')
```

```python
df['Last_Name']=df['Last_Name'].str.strip('/')
df['Last_Name']=df['Last_Name'].str.strip('...')
df['Last_Name']=df['Last_Name'].str.strip('_')
df
```

```python
df['Last_Name'].str.strip('123./_')
#123 is like assigning a variale to the list
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]','',regex=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:x[0:4]+'-'+x[4:10])
df
```

```python
df[['street_address', 'state','zip-code']]=df['Address'].str.split(',',n=2, expand =True)
#n = number of columns
df.drop(columns=['Address'], inplace = True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df['Paying Customer']= df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df.fillna('')
```

```python
df=df.replace('N/a', '')
df
```

```python
df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='':
        df.drop(x, inplace=True)
df
```

```python
df.reset_index(drop=True)
```

```python
df.to_excel('E:\DAprojects', index= False)
```
